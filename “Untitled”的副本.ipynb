```python
import numpy as np
n,p,m=2,3,4
A=np.random.randint(0,10,(n,p))
B=np.random.randint(0,10,(p,m))
dot_product=np.dot(A,B)
print("\nMatrixA({}*{}:\n".format(n,p),A)
print("\nMatrixB({}*{}:\n".format(p,m),B)
print("\nDot Product ({}*{}):\n".format(n,m),dot_product)
```
